### Importing the necessary packages:

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
plt.rcParams['figure.figsize'] = (10,8) #For figure appearance 
```

### A figure of the Antarctic and the Southern Ocean:
Using a South Polar stereographic projection. This projection is best because it displays the Antarctic continent with the least distortion possible. 

```python
fig = plt.figure(figsize=(8,8))

ax = plt.axes(projection=ccrs.SouthPolarStereo())

ax.set_extent([-180, 180, -90, -60], crs=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND, facecolor='#d9d9d9')
ax.add_feature(cfeature.OCEAN, facecolor='#c6dbef')
ax.coastlines(resolution='50m', linewidth=1)

gl = ax.gridlines(draw_labels=True, linestyle='--')
gl.top_labels = False
gl.right_labels = False

plt.title('Antarctica and the Southern Ocean (South of 60°S)', fontsize=14)

ax.spines['geo'].set_linewidth(1) # Adding borders to the maps 

plt.show()
```

### Figure 2: The South Atlantic (20ºS to 50ºS)
Using the Mercator projection because of how it preserves the shapes of the mid-latitudes continental and ocean shapes. 
The dark green representing the land and the light blue being the ocean. 

```python
fig = plt.figure(figsize=(10,8))

ax = plt.axes(projection=ccrs.Mercator())

ax.set_extent([-70, 20, -50, -20], crs=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND, facecolor='darkgreen')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
ax.coastlines(resolution='50m')

gl = ax.gridlines(
    draw_labels=True,
    linewidth=0.6,
    color='gray',
    alpha=0.6,
    linestyle='--'
)

gl.top_labels = False
gl.right_labels = False
gl.xlabel_style = {'size': 10}
gl.ylabel_style = {'size': 10}

# City coordinates
cities = {
    "Walvis Bay": (14.505, -22.957),
    "Cape Town": (18.424, -33.925),
    "Rio de Janeiro": (-43.172, -22.906),
    "Montevideo": (-56.164, -34.901)
}

for city, (lon, lat) in cities.items():
    ax.plot(lon, lat, 'ro', transform=ccrs.PlateCarree())

    ax.text(lon+1, lat, city,
            transform=ccrs.PlateCarree(),
            fontsize=10,
            color='black',
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3')) #making the city labels black text with white box to appear more legible

ax.text(-65, -48, 'Scale ~1000 km',
        transform=ccrs.PlateCarree(),
        fontsize=9) # Adding a scale bar

plt.title('South Atlantic Ocean (20°S – 50°S)', fontsize=14)

ax.spines['geo'].set_linewidth(1)

plt.show()
```

### Figure 3: Three panels each showing varying levels of resolution of False Bay, Western Cape, South Africa. 
The panels are using a mercator projection as it is a small area and the shape of the coastline is the most accurate in the Mercator projection. 

```python
fig, axes = plt.subplots(
    1, 3,
    figsize=(15,6),
    subplot_kw={'projection': ccrs.Mercator()}
)

resolutions = ['110m', '50m', '10m']
titles = ['Coarse Resolution', 'Intermediate Resolution', 'Full Resolution']

for ax, res, title in zip(axes, resolutions, titles):

    ax.set_extent([18.2, 19.2, -34.6, -33.8], crs=ccrs.PlateCarree())

    ax.add_feature(cfeature.OCEAN, facecolor='lightblue')

    ax.coastlines(resolution=res, linewidth=1.8)

    gl = ax.gridlines(draw_labels=True, linestyle='--')
    gl.top_labels = False
    gl.right_labels = False

    ax.set_title(title, fontsize=12)

plt.suptitle('Coastline Resolution Comparison in False Bay', fontsize=16)

ax.spines['geo'].set_linewidth(1)

plt.tight_layout()
plt.show()
```
